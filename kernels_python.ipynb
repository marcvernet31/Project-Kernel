```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
import sklearn
from sklearn.model_selection import train_test_split as tts

from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score
```

```python
class Dataset:
    def __init__(self, *args, **kwargs):
        self.data_arrays = self.generate()

    def generate(self):
        raise NotImplementedError

    def train_test_split(self, testSize = 0.3, randomState = 0):
        return tts(*self.data_arrays, test_size=testSize, random_state=randomState)

class CannabisGenotype(Dataset):
    def generate(self):
        X, y = [], []
        with open("data/cannabis_gen.csv", "r") as data:
            head = True
            for line in data:
                if not head:
                    seq = line.split(',')
                    y.append(int(seq[-1][1:-2]))
                    X.append([int(x.strip()) for x in seq[1:-1]])
                else:
                    head = False
        X, y = np.array(X), np.array(y)
        return X, y
```

```python
data = CannabisGenotype()
X_train, X_test, y_train, y_test = data.train_test_split()
```

```python
# CUSTOM KERNELS

    

def k0_exp(X, Y):
    xm, xn = X.shape
    ym, yn = Y.shape
    # Compute the kernel matrix:
    G = np.zeros((xm, ym))
    for i in range(xm):
        Xi = np.tile(X[i], (ym, 1))
        Xi = np.exp(Xi == Y) # prevf f1
        G[i, :] = np.sum(Xi, axis=1)
    return G # postf f2 f2(G)

def k0_sqrt_p(X, Y):
    xm, xn = X.shape
    ym, yn = Y.shape
    # Compute the kernel matrix:
    G = np.zeros((xm, ym))
    for i in range(xm):
        Xi = np.tile(X[i], (ym, 1))
        Xi = (Xi == Y) # prevf f1
        G[i, :] = np.sqrt(np.sum(Xi, axis=1))
    return G # postf f2 f2(G)

def k0_sqrt(X, Y):
    xm, xn = X.shape
    ym, yn = Y.shape
    # Compute the kernel matrix:
    G = np.zeros((xm, ym))
    for i in range(xm):
        Xi = np.tile(X[i], (ym, 1))
        Xi = np.sqrt(Xi == Y) # prevf f1
        G[i, :] = np.sum(Xi, axis=1)
    return G # postf f2 f2(G)

# K0 (overlap)
def k0(X, Y):
    gamma = 1
    xm, xn = X.shape
    ym, yn = Y.shape
    # Compute the kernel matrix:
    G = np.zeros((xm, ym))
    for i in range(xm):
        Xi = np.tile(X[i], (ym, 1))
        Xi = Xi == Y # prevf
        G[i, :] = np.sum(Xi, axis=1)
    return G # postf f(G)

def jaccard (X,Y):
    xm, xn = X.shape
    ym, yn = Y.shape
    n = len(X[0])
    # Compute the kernel matrix:
    G = np.zeros((xm, ym))
    for i in range(xm):
        Xi = np.tile(X[i], (ym, 1))
        Xi = Xi == Y # prevf
        eq = np.sum(Xi, axis=1)
        G[i, :] = eq/(2*n - eq)
    return G # postf f(G)
    
```

```python
class Dataset:
    def __init__(self, *args, **kwargs):
        self.data_arrays = self.generate()

    def generate(self):
        raise NotImplementedError

    def train_test_split(self, testSize = 0.3, randomState = 0):
        return tts(*self.data_arrays, test_size=testSize, random_state=randomState)

class CannabisGenotype(Dataset):
    def generate(self):
        X, y = [], []
        with open("data/original_encoding.csv", "r") as data:
            head = True
            for line in data:
                if not head:
                    seq = line.split(',')
                    y.append(int(seq[-1][1:-2]))
                    X.append([int(x.strip()) for x in seq[1:-1]])
                else:
                    head = False
        X, y = np.array(X), np.array(y)
        return X, y

original = CannabisGenotype()
X_train, X_test, y_train, y_test = original.train_test_split()
X, y = original.generate()
```

```python
def SMC(X, Y):
    xm, xn = X.shape
    ym, yn = Y.shape
    # Compute the kernel matrix:
    G = np.zeros((xm, ym))
    for i in range(xm):
        Xi = np.tile(X[i], (ym, 1))
        Xi = Xi == Y # prevf
        G[i, :] = (np.sum(Xi, axis=1)/xn)
    return G # postf f(G)

def iguals(X, Y):
    X = X[0]
    Y = Y[0]
    n = len(X)
    eq = 0
    for i in range(0, n-1, 2):
        eq += (X[i] == X[i+1]) == (Y[i] == Y[i+1])
    return eq/(n/2)

def SMC_creuat(X,Y):
    X = X[0]
    Y = Y[0]
    n = len(X)
    eq = 0
    for i in range(0, n-1, 2):
        eq += (X[i] == Y[i+1]) 
        eq += (Y[i] == X[i+1])
    return eq/n

def prova(X, Y):
    gamma = 1
    xm, xn = X.shape
    ym, yn = Y.shape
    # Compute the kernel matrix:
    G = np.zeros((xm, ym))
    for i in range(xm):
        Xi = np.tile(X[i], (ym, 1))
        Xi = (Xi == Y) # prevf
        #G[i, :] = np.exp(gamma * (np.sum(Xi, axis=1)/xn))
        G[i, :] = 0.8*(np.sum(Xi, axis=1)/xn) + 0.1*iguals(X,Y) + 0.1*SMC_creuat(X,Y)

    return G # postf f(G)
```

```python
df = one_hot

if 'family' in df.columns:
    df_matrix = df.drop('ID', axis=1).drop('drug', axis=1).drop('family', axis=1)
    
else:
    df_matrix = df.drop('ID', axis=1).drop('drug', axis=1)
    
df_matrix = df_matrix.to_numpy()

size = 100

kernel_one_hot = np.empty([size, size])
for i in range(0, size):
    for j in range(0, size):
        X = df_matrix[i].reshape(1, -1)
        Y = df_matrix[j].reshape(1, -1)
        kernel_one_hot[i][j] = float(prova(X,Y))

original = pd.read_csv('data/original_encoding.csv')
df_matrix = original.to_numpy()

kernel_original = np.empty([size, size])
for i in range(0, size):
    for j in range(0, size):
        X = df_matrix[i].reshape(1, -1)
        Y = df_matrix[j].reshape(1, -1)
        kernel_original[i][j] = float(prova(X,Y))

plt.title('original')
plt.imshow(abs(kernel_original), cmap='hot')
plt.show()
 
plt.title('one_hot')
plt.imshow(abs(kernel_one_hot), cmap='hot')
plt.show()

plt.title('difference')
plt.imshow(abs(kernel_one_hot - kernel_original), cmap='hot')
plt.show()
```
