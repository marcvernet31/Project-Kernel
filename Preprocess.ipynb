```python
import pandas as pd
import numpy as np
```

```python
column_names = ["ID", "A201_1","A201_2",
                "A301_1","A301_2",
                "A303_1","A303_2",
                "A305_1","A305_2",
                "P15_1","P15_2",
                "A202_1","A202_2",
                "A302_1","A302_2",
                "A501_1","A501_2",
                "C11_C1_1","C11_C1_2",
                "B02_C2_1","B02_C2_2",
                "A304_1","A304_2",
                "B05_C1_1","B05_C1_2",
                "B01_C1_1","B01_C1_2"]
raw = pd.read_csv("data/CannabisGenotypes.csv", names = column_names, header = 0, dtype = "object")
raw.head()
```

```python
labeled = raw
for i in range(341):
    ide = "ID"
    if i <= 99:
        ide = ide + "0"
    if i <=9:
        ide = ide + "0"
    ide = ide + str(i)
    labeled = labeled[labeled["ID"] != ide]
labeled = labeled[labeled["ID"] != "ID"]
```

### Add the label of drug or nondrug

```python
cannabis_gen = pd.read_csv('data/cannabis_gen.csv')
labeled["drug"] = cannabis_gen["drug"]
```

### Add a variable indicating the family

```python
family = list(labeled['ID'])
for i in range(0, len(family)):
    family[i] = family[i].split('_')[0]
labeled['family'] = family
```

```python
labeled.head()
```

```python
order = ["ID", "family", "drug"] + column_names[1:]
labeled = labeled[order]
labeled
```

```python
labeled.to_csv("data/labeledGenotype.csv", na_rep = -99)
```
